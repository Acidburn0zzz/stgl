```python
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import numpy.polynomial.chebyshev as chebyshev
from itertools import *
```

```python
img = Image.open('./flame.png')
colors = np.array([img.getpixel((i, 0)) for i in range(img.size[0])])
r, g, b = np.array_split(colors / 255, 3, axis=1)

plt.plot(r, 'r')
plt.plot(g, 'g')
plt.plot(b, 'b')
```

```python
xs = np.linspace(0, 1, len(r))
approx_ord = 9;
cr = chebyshev.chebfit(xs, r, approx_ord)
cg = chebyshev.chebfit(xs, g, approx_ord)
cb = chebyshev.chebfit(xs, b, approx_ord)
"""

plt.plot(r, 'r--')
plt.plot(g, 'g--')
plt.plot(b, 'b--')

plt.figure();

plt.plot(chebyshev.chebval(xs, cr) - r, 'r')
plt.plot(chebyshev.chebval(xs, cg) - g, 'g')
plt.plot(chebyshev.chebval(xs, cb) - b, 'b')
"""
plt.plot(chebyshev.chebval(xs, cr), 'r')
plt.plot(chebyshev.chebval(xs, cg), 'g')
plt.plot(chebyshev.chebval(xs, cb), 'b')

```
